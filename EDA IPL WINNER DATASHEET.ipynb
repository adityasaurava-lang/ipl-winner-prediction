There are many columns which is not used for prediction so you could remove it

```python
data['is_wicket'] = data['player_dismissed'].notnull().astype(int)
```

```python
keep_columns = [
    'match_id','inning','batting_team','bowling_team',
    'over','ball','total_runs','is_wicket',
    'winner','venue'
]

data = data[keep_columns]
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
#2nd innings
data=data.query('inning==2')
```

```python
# 1
temp = deliveries.merge(matches, left_on='match_id', right_on='id')
temp = temp[temp['inning'] == 1]

# 2
target_df = temp.groupby('match_id')['total_runs'].sum().reset_index()
target_df['target'] = target_df['total_runs'] + 1

# 3
data = data.merge(target_df[['match_id','target']], on='match_id')

# 4
data['runs_left'] = data['target'] - data['total_runs']
```

```python
#wickets left
data['wickets'] = data.groupby('match_id')['is_wicket'].cumsum()
data['wickets_left'] = 10 - data['wickets']
```

```python
#ball_left_columns
data['balls_left'] = 120 - (data['over']*6 + data['ball'])
```

```python
#required run rate
data['rrr'] = (data['runs_left'] * 6) / data['balls_left']
```

```python
data.columns
```

```python
data.head()
```

```python
data['result'] = (data['batting_team'] == data['winner']).astype(int)
```

```python
import numpy as np

# infinity to nan
data.replace([np.inf, -np.inf], np.nan, inplace=True)

# nan is removed
data = data.dropna()
```

```python
#final features
```

```python
#targeting columns for modals
features = [
    'batting_team',
    'bowling_team',
    'venue',
    'runs_left',
    'balls_left',
    'wickets_left',
    'rrr'
]

X = data[features]
y = data['result']
```

```python
#encoding
X = pd.get_dummies(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)
```

```python
#accuracy
print(model.score(X_test, y_test))
```

```python
import pickle

pickle.dump(model, open('ipl_winning.pkl', 'wb'))
pickle.dump(X.columns, open('columns.pkl', 'wb'))
```
